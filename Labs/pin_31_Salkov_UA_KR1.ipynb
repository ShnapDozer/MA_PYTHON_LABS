```python
from sympy import symbols, lambdify, sin, pi

from sympy import *
import matplotlib.pyplot as plt

def up_sums(f, a, b, n):
    sd = 0
    gap = (b - a)/n
    
    for i in range(0, n):
        sd += maximum(f, x, Interval(a + gap*i, a + gap*(i+1))) * gap;
    
    print("Верхние суммы Дабру вычесленно: ", sd, "При n =", n)
    
def down_sums(f, a, b, n):
    s1 = 0;
    sd = 0
    
    gap = (b - a)/n
    
    for i in range(0, n):
        sd += minimum(f, x, Interval(a + gap*i, a + gap*(i+1))) * gap;

    print("Нижние суммы Дабру вычесленно: ", sd, "При n =", n)

x = symbols('x')
n=1

f = x**2 - x + 2
a = 0
b = 2
up_sums(f, a, b, n)

f = -2*x
a = 2
b = 4
up_sums(f, a, b, n)

f = x**2 - x + 2
a = 0
b = 2
down_sums(f, a, b, n)

f = -2*x
a = 2
b = 4
down_sums(f, a, b, n)
```
