Построить графики функций. Найти точки пересечения графиков. Вычислить площадь фигуры, ограниченной графиками.

```python
from sympy import *
from sympy.plotting import plot

x = symbols('x')

func1 = (x**2 + 2*x)
func2 = -x**2 -4*x +7

plot(func1, func2)
commonPoints = solve(2*x**2 + 6*x - 7, x)

pprint(integrate(func1 - func2,(x, commonPoints[0], commonPoints[1])))
```

## 4. Вычисление длины дуги.

### Упражнение 5. 
Найти длину дуги параболы от точки(1;1)A до точки(2;4).

```python
from sympy import *

x = symbols('x')
func = x**2
print("Результат: ", simplify(integrate(sqrt(1+(func.diff(x)**2)),(x, 1, 4))))
```

### Упражнение 6. 
Найти длину замкнутой кривой, заданной параметрическими уравнениями.

```python
import scipy.integrate as integrate
import numpy as np

t = np.linspace(0, 2*np.pi,1000)

dFuncX = -2*np.sin(t)
dFuncY = 3*np.cos(t)

print("Результат: ", integrate.simpson(np.sqrt(dFuncX**2 + dFuncY**2),t))
```

## 5. Вычисление объема тела вращения.

### Упражнение 8. 
Вычислить объем тела, полученного при вращении криволинейной трапеции, ограниченной графиком функции и прямой

    а) относительно оси Ox 
    б) относительно оси Oy

```python
from sympy import *

x = symbols('x')
func = sin(x)
print("Результат Ox: ", simplify(integrate(func**2,(x, 0, pi)) * pi))
print("Результат Oy: ", simplify(integrate(x * func,(x, 0, pi)) * 2*pi))
```
