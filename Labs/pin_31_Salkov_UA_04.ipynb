```python
import math

def euclidean_distance(v1, v2):
    return math.sqrt(sum([(x - y) ** 2 for x, y in zip(v1, v2)]))

def manhattan_distance(v1, v2):
    return sum([abs(x - y) for x, y in zip(v1, v2)])

def chebyshev_distance(v1, v2):
    return max([abs(x - y) for x, y in zip(v1, v2)])

# Проверка работы функций
print(euclidean_distance([1, 2, 3, 4], [7, 3, 4, 200]))
print(manhattan_distance([1, 2, 3, 4], [7, 3, 4, 200])) 
print(chebyshev_distance([1, 2, 3, 4], [7, 3, 4, 200])) 
```

### Упражнение 2.
Создать M-функцию, строящую изображение замкнутого шара в2 R для различных метрик. Построить шары,1OB в метриках0 ,1,2

```python
import matplotlib.pyplot as plt
import numpy as np

def euclidean_ball(center, radius):
    x = np.linspace(center[0]-radius, center[0]+radius, 1000)
    y = np.linspace(center[1]-radius, center[1]+radius, 1000)
    X, Y = np.meshgrid(x, y)
    F = (X-center[0])**2 + (Y-center[1])**2 - radius**2
    plt.contour(X, Y, F, [0], colors='r')
    plt.axis('equal')
    plt.show()
    
def manhattan_ball(center, radius):
    x = np.linspace(center[0]-radius, center[0]+radius, 1000)
    y = np.linspace(center[1]-radius, center[1]+radius, 1000)
    
    X, Y = np.meshgrid(x, y)
    
    F = np.abs(X-center[0]) + np.abs(Y-center[1]) - radius
    
    plt.contour(X, Y, F, [0], colors='r')
    plt.axis('equal')
    plt.show()

def chebyshev_ball(center, radius):
    x = np.linspace(center[0]-radius, center[0]+radius, 10000)
    y = np.linspace(center[1]-radius, center[1]+radius, 10000)
    X, Y = np.meshgrid(x, y)
    F = np.maximum(np.abs(X-center[0]), np.abs(Y-center[1])) - radius
    plt.contour(X, Y, F, [0], colors='r')
    plt.axis('equal')
    plt.show()

# Пример использования функции
euclidean_ball(center=[0, 0], radius=1)
manhattan_ball(center=[0, 0], radius=1)
chebyshev_ball(center=[0, 0], radius=1)
```

```python
import numpy as np
 
def sort_diag_arr(a,b) :
    n = len(a)
    a = np.array(a)
    b = np.array(b)
    for i in range(n):
        ind = np.abs(a[i:, i:n]).argmax()
        row = ind // (n - i) + i
        col = ind % (n - i) + i 
        a[:,[i, col]] = a[:,[col, i]]
        a[[i, row], :] = a[[row, i], :]
        b[i], b[row] = b[row], b[i]
    return a, b
 
 
A = [[0.23,-0.29, 5.41], [8.31 ,-0.24, 7.27], [-0.41,-0.23, -0.67]]
B = [-4.26,7.51,6.23]
 
eps = 1e-3
 
a, b = sort_diag_arr(A,B)
diag = (1/np.diag(a)).reshape(-1,1)
a[np.diag_indices_from(a)] = 0.
a = np.hstack((-a, b.reshape(-1, 1))) * diag
x = a[:,-1].ravel()
x = np.append(x, 1)
 
tmp = x.copy() + eps
cnt = 0
while abs(x-tmp).sum() > eps:
    tmp = x.copy()
    x = a.dot(x)
    x = np.append(x, 1)
    cnt += 1
 
print(x[:-1].round(3), cnt)
```

```python
import numpy as np

# задаем коэффициенты системы уравнений
A = np.array([[0.23,-0.29, 5.41], [8.31 ,-0.24, 7.27], [-0.41,-0.23, -0.67]])
b = np.array([-4.26,7.51,6.23])

# задаем начальное приближение x0
x0 = np.zeros(len(b))

# задаем максимальное число итераций и желаемую точность решения
max_iterations = 100
tolerance = 1e-6

# выполняем итерации
for i in range(max_iterations):
    x1 = np.zeros(len(b))
    for j in range(len(b)):
        s = 0
        for k in range(len(b)):
            if k != j:
                s += A[j,k] * x0[k]
        x1[j] = (b[j] - s) / A[j,j]
    if np.allclose(x0, x1, rtol=tolerance):
        break
    x0 = x1

# выводим решение
print("Solution:", x1)
```
