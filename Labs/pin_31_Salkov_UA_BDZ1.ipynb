## 1. Найти площадь, ограниченную кривыми

```python
from sympy import *
from sympy.plotting import plot

x = symbols('x')

func = x - (x**2)*sqrt(x)
plot(func, (x, 0, 1))

pprint(integrate(func,limits))
```

## 2. Найти площадь, ограниченную кривыми

```python
from numpy import *
import matplotlib.pyplot as plt
t=linspace(0,2*pi,1000)

r1=6*sin(t)
x1=r1*cos(t)+1
y1=r1*sin(t)

r2=4*sin(t)
x2=r2*cos(t)+1
y2=r2*sin(t)

plt.plot(x1,y1)
plt.plot(x2,y2)
plt.grid(True)
```

```python
from sympy import *
from sympy.plotting import plot

p = symbols('p')

r1 = 6*sin(p)
r2 = 4*sin(p)

s = 1/2 * integrate(r1**2,(p, 0, 2*pi)) - 1/2 * integrate(r2**2,(p, 0, 2*pi))

pprint(s)
```

## 3. Вычислить длину дуги

```python
from sympy import *

x = symbols('x')

func = ln(x)
plot(func, (x, sqrt(3), sqrt(15)))
print("Результат: ", simplify(integrate(sqrt(1+(func.diff(x)**2)),(x, sqrt(3), sqrt(15)))))
```

## 5. Вычислить длину дуги, заданую в полярной системе координат

```python
from numpy import *
import matplotlib.pyplot as plt

t=linspace(0,2*pi,1000)

r = 2*(1-cos(t))

x=r*cos(t)+1
y=r*sin(t)

plt.plot(x,y)
plt.grid(True)
```

```python
import scipy.integrate as integrate
import numpy as np

t = np.linspace(0, 2*np.pi,1000)

Func = 2*(1-np.cos(t))
dFunc = 2*np.sin(t)

print("Результат: ", integrate.simpson(np.sqrt(Func**2 + dFunc**2),t))
```
