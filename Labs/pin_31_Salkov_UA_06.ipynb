## 1. Функции многих переменных. Область определения.
### Упражнение 1.
Создать функции, строящие графики функций y=f(x), y=g(x) на промежутке [a,b] (f(x) <= g(x), при x принадлежащему орезку [a,b]) и показывающую вертикальной штриховкой область, заключенную между графиками. Входные аргументы – функции, границы отрезка [a,b] и количество частей, на  оторые разбивается отрезок:

    а) верхняя и нижняя границы сплошные;
    б) обе границы пунктирные;
    в) одна граница сплошная, другая пунктирная.

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_functions_mixed(f, g, a, b, n, linestyle_a='solid', linestyle_b='dashed'):
    x = np.linspace(a, b, n)
    y1 = f(x)
    y2 = g(x)
    plt.plot(x, y1, label='f(x)', linestyle=linestyle_a)
    plt.plot(x, y2, label='g(x)', linestyle=linestyle_b)
    plt.fill_between(x, y1, y2, where=y1<=y2, interpolate=True, alpha=0.5)
    plt.legend()
    plt.show()
    

    
f = lambda x: x**2
g = lambda x: x**3

a = -5
b = 5
n = 100

plot_functions_mixed(f, g, a, b, n, linestyle_a='solid', linestyle_b='solid')


    
f = lambda x: x**2
g = lambda x: x**3

a = -5
b = 5
n = 100

plot_functions_mixed(f, g, a, b, n, linestyle_a='dashed', linestyle_b='dashed')


    
f = lambda x: x**2
g = lambda x: x**3

a = -5
b = 5
n = 100

plot_functions_mixed(f, g, a, b, n, linestyle_a='solid', linestyle_b='dashed')
```

Используя М-функции из упр. 1, построить области определения функции

```python
f = lambda x: x**2 - 1;
g = lambda x: x**2 + 1;

plot_functions_mixed(f, g, a, b, n, linestyle_a='solid', linestyle_b='solid')
plot_functions_mixed(f, g, a, b, n, linestyle_a='dashed', linestyle_b='dashed')
plot_functions_mixed(f, g, a, b, n, linestyle_a='solid', linestyle_b='dashed')
```

### Пример 1.
Сгенерируем сетку на области[ 1;1],x  [0;1]y  и значения в узлах для функции  4sin 2 cos1.5 (1 ) (1 ) :

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(projection = '3d')

xs = np.linspace(-1, 1, 50)
ys = np.linspace(-1, 1, 50)

X, Y = np.meshgrid(xs, ys)
Z = 4*np.sin(2*np.pi*X)*np.cos(1.5*Y)*(1-X**2)*Y*(1-Y)

ax.plot_wireframe(X, Y, Z, rstride = 2, cstride = 2)
```

### Упражнение 1С.
Построить области определения следующих функций:

```python
import numpy as np
import matplotlib.pyplot as plt

# Определяем диапазон значений x и y
x = np.linspace(-1, 1, 100)
y = np.linspace(0, 1, 100)

# Создаем сетку координат x и y
X, Y = np.meshgrid(x, y)

# Вычисляем значения функции f(x,y) = ln(1 - x**2 - y**2)*ln(y)
Z = np.log(1 - X**2 - Y**2) * np.log(Y)

# Строим область определения функции f(x,y) = ln(1 - x**2 - y**2)*ln(y)
plt.imshow(Z, extent=[-1, 1, 0, 1])

plt.title('Область определения функции f(x,y) = ln(1 - x**2 - y**2)*ln(y)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Определяем диапазон значений x и y
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)

# Создаем сетку координат x и y
X, Y = np.meshgrid(x, y)

# Вычисляем значения функции f(x,y) = sqrt((4-x**2 - y**2) / (x**2 + y**2 -1))
Z = np.sqrt((4 - X**2 - Y**2) / (X**2 + Y**2 - 1))

# Строим область определения функции f(x,y) = sqrt((4-x**2 - y**2) / (x**2 + y**2 -1))
plt.imshow(Z, extent=[-3, 3, -3, 3], cmap='inferno')
plt.colorbar()
plt.title('Область определения функции f(x,y) = sqrt((4-x**2 - y**2) / (x**2 + y**2 -1))')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
